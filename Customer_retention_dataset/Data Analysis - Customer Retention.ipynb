### Data Cleaning and Manipulation

```python
# printing all columns with the total no. of nulls present in each column.
obj = df.isnull().sum()
count = 0
for key,value in obj.items():
    if value > 0:
        print(key,",  ",value)
        count =+ 1
print (f"{count} columns have null values in it.")
```

Our dataset has no null values. So, let's proceed.

Let's check the consistency of the responses.

```python
for i in df.columns:
    print(df.groupby ([i]).size())
    print('-'*25, '*****', '-'*25)
```

In "Q6 How many times you have made an online purchase in the past 1 year?"\
we can see there is inconsistency in data as it should be 41 times and above for the responses having 42 times and above. Since have both "41 times and above" & "42 times and above" doesn't make any sense.

In "Q7 How do you access the internet while shopping on-line?"
We can see there is inconsistency in the "Mobile Internet" and "Mobile internet" enteries. Since both are the same so having them spilt counts needs to be corrected.

So, let's replace the "42 times and above" with "41 times and above".\
So, let's replace the "Mobile internet" with "Mobile Internet".

```python
df = df.replace('42 times and above', '41 times and above')
df = df.replace('Mobile internet', 'Mobile Internet')
for i in df.columns:
    print(df.groupby ([i]).size())
    print('-'*25, '*****', '-'*25)
```

Now everything looks clean. Let's sort options for each question in descending.

```python
for i in df.columns:
    print(df.groupby ([i]).size().sort_values(ascending=False))
    print('-'*25, '*****', '-'*25)
```

Now let's see graphical representation of the same.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total = float(len(df))

for i in df.columns:
    if i == '4 What is the Pin Code of where you shop online from?':
        #Q4 was skipped in subplots as it had many bars and annotations and labeling got unbalanced compared to all
        pass
    else:
        fig, ax = plt.subplots(1,1,figsize = (15,6))
        sns.countplot(y = df[i][:],data=df.iloc[:], order=df[i][:].value_counts().index)
        fig.text(0.1, 0.95, f'{i}', fontsize=16, fontweight='bold', fontfamily='serif')
        plt.xlabel(' ', fontsize=20)
        plt.ylabel('')
        plt.yticks(fontsize=13)
        plt.box(False)
        for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_width()/total)
            y = p.get_y() + p.get_height()
            x = p.get_width()
            ax.annotate(percentage, (x, y), ha ='left', size = 30)
```
