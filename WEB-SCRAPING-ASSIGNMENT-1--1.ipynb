```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1) Write a python program to display all the header tags from wikipedia.org.
def headers(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    header = [] # empty list to store all headers

    for i in soup.find_all(['h1','h2','h3','h4','h5','h6']): #since there are total 6 headers in html we pass them as list
        header.append(i)
    return (header)
headers('https://en.wikipedia.org/wiki/Main_Page')
```

```python
#2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

def movies(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    
    movie = [] # empty list to store all movies
    for i in soup.find_all('h3', class_="lister-item-header"):
        movie.append(i.text.split("\n")[2])
        
    rating = [] # empty list to store all ratings
    for i in soup.find_all('div', class_ = "ipl-rating-star small"):
        rating.append(i.text.split('\n')[8])
    
    year = [] # empty list to store all years
    for i in soup.find_all('span', class_="lister-item-year text-muted unbold"):
        year.append(i.text.split('(',)[1].split(')')[0])
        
    df = pd.DataFrame({'Movies':movie, 'Ratings':rating, 'Year of Release':year})
    df
    return (df)

# calling the function using link as the parameter
movies('https://www.imdb.com/list/ls091520106/')
```

```python
# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

def movies(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    
    movie = [] # empty list to store all movies
    for i in soup.find_all('td', class_="titleColumn"):
        movie.append(i.text.split("\n")[2].strip())
    del movie[100:250]
        
    rating = [] # empty list to store all ratings
    for i in soup.find_all('td', class_ = "ratingColumn imdbRating"):
        rating.append(i.text.split("\n")[1])
    del rating[100:250]
    
    year = [] # empty list to store all years
    for i in soup.find_all('td', class_="titleColumn"):
        year.append(i.text.split("\n")[3].split('(',)[1].split(')')[0])
    del year[100:250]
        
    df = pd.DataFrame({'Movies':movie, 'Ratings':rating, 'Year of Release':year})
    df
    return (df)
# calling the function using link as the parameter
movies("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
'''
4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office)
from https://presidentofindia.nic.in/former-presidents.htm
'''

def indian_president(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    
    president = [] # empty list to store former president names
    for i in soup.find_all('h3'):
        president.append(i.text.split("(")[0].strip())
        
    term_of_office = [] # empty list to store all term of office
    for i in soup.find_all('p'):
        term_of_office.append(i.text)
    term_of_office
    filter_='Term'
    term = []
    for i in term_of_office:
        if(i.find(filter_)==0):
            term.append(i.split(":")[1].strip())
        
    df = pd.DataFrame({'Former President':president, 'Term of office':term})
    df
    return (df)

# calling the function using link as the parameter
indian_president("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
'''
5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
'''

def top_team(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    
    team = [] # empty list to store all teams
    for i in soup.find_all('span', class_="u-hide-phablet"):
        team.append(i.text)
        
    matches = [] # empty list to store all matches
    for i in soup.find_all('tr', class_ = "rankings-block__banner"):
        matches.append(i.text.split("\n")[7].strip())
    for i in soup.find_all('tr', class_ = "table-body"):
        matches.append(i.text.split("\n")[7].strip())
    
    points = [] # empty list to store all points for teams
    for i in soup.find_all('tr', class_ = "rankings-block__banner"):
        points.append(i.text.split("\n")[8].strip())
    for i in soup.find_all('tr', class_ = "table-body"):
        points.append(i.text.split("\n")[8].strip())
            
    rating = [] # empty list to store all ratings
    for i in soup.find_all('tr', class_ = "rankings-block__banner"):
        rating.append(i.text.split("\n")[10].strip())
    for i in soup.find_all('tr', class_ = "table-body"):
        rating.append(i.text.split("\n")[9].strip())
        
    df = pd.DataFrame({'Team':team, 'Matches':matches, 'Points':points, 'Rating':rating})
    df
    return (df)

```

```python
# calling the function using link as the parameter
top_team_men = top_team('https://www.icc-cricket.com/rankings/mens/team-rankings/test')
top_team_men
```

```python
'''
5.) b) Top 10 ODI Batsmen along with the records of their team and rating.
'''

def top_ten(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)
    
    player = [] # empty list to store top 10 odi batsmen
    for i in soup.find_all('div', class_ = "rankings-block__banner--name-large"):
        player.append(i.text)
    for i in soup.find_all('td', class_ ="table-body__cell rankings-table__name name"):
        player.append(i.text.split("\n")[1])
    del player[10:100]
    
    team = [] # empty list to store top 10 odi batsmen's team
    for i in soup.find_all('div', class_ = "rankings-block__banner--nationality"):
        team.append(i.text.split("\n")[2])
    for i in soup.find_all('span', class_ ="table-body__logo-text"):
        team.append(i.text)
    del team[10:100]
    
    rating = [] # empty list to store top 10 odi batsmen's rating
    for i in soup.find_all('div', class_ = "rankings-block__banner--rating"):
        rating.append(i.text)
    for i in soup.find_all('td', class_ ="table-body__cell rating"):
        rating.append(i.text)
    del rating[10:100]
        
    df = pd.DataFrame({'Player Name':player, 'Team':team, 'Rating':rating})
    df
    return (df)

```

```python
# calling the function using link as the parameter
top_ten_batsmen_men = top_ten("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
top_ten_batsmen_men
```

```python
'''
7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
'''
def headline(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)

    news = []
    for i in soup.find_all('div', class_ = "LatestNews-headlineWrapper"):
        news.append(i.text.replace("Ago", ",", 1).replace("2022", " ", 1).split(",")[1].strip())
        
    time = []
    for i in soup.find_all('span', class_ = "LatestNews-wrapper"):
        time.append(i.text)
        
    df = pd.DataFrame({'Headline':news, 'Time':time})
    df
    return (df)

headline("https://www.cnbc.com/world/?region=world")
```

```python
'''
8) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details :

'''
def journal(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)

    # i) Paper Title
    paper_title = []
    for i in soup.find_all('h2', class_ = "sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR"):
        paper_title.append(i.text)
    
    # 8.) ii) Authors
    authors = []
    for i in soup.find_all('span', class_ = "sc-1w3fpd7-0 pgLAT"):
        authors.append(i.text.strip())
    
    # 8.) iii) Published Date

    date = []
    for i in soup.find_all('span', class_ = "sc-1thf9ly-2 bKddwo"):
        date.append(i.text)
        
    journals = pd.DataFrame({'Paper Title':paper_title, 'Authors':authors, 'Published Date': date})
    return(journals)

journal("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
```

```python
'''
9) Write a python program to scrape mentioned details from dineout.co.in :

'''

def restaurant(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)

    # i) Restaurant name
    restaurant_name = []
    for i in soup.find_all('a', class_="restnt-name ellipsis"):
        restaurant_name.append(i.text)
    restaurant_name
    
    # 9.) ii) Cuisine

    cuisine = []
    for i in soup.find_all('span', class_ = 'double-line-ellipsis'):
        cuisine.append(i.text.split('|')[1].strip())
        
    # 9.) iii) Location

    location = []
    for i in soup.find_all('div', class_ = 'restnt-loc ellipsis'):
        location.append(i.text)  

    # 9.) iv) Ratings

    rating = []
    for i in soup.find_all('div', class_ = "restnt-rating rating-4"):
        rating.append(i.text)
    
    # 10.) v) Image URL

    # Extracting links was not covered in the last web scrapping session.
    # But I'm attempting this as to what was covered in the last session.

    images = []
    for i in soup.find_all("img", class_="no-img"):
        images.append(i.get('data-src'))

    restau = pd.DataFrame({'Restaurant name':restaurant_name, 'Cuisine':cuisine, 'Location':location, 'Ratings':rating, 'Image URL': images})
    return(restau)

restaurant("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
'''
10) Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank
'''

def pub(x):
    page = requests.get(x)
    soup = BeautifulSoup(page.content)

    rank = []
    for i in soup.find_all('td', class_="gsc_mvt_p"):
        rank.append(i.text)

    # 10.) ii) Publication

    publication = []
    for i in soup.find_all('td', class_="gsc_mvt_t"):
        publication.append(i.text)
        
    # 10.) iii) h5-index

    temp = []     # empty list to fetch all text for class="gsc_mvt_n"
    for i in soup.find_all('td', class_="gsc_mvt_n"):
        temp.append(i.text)

    h5_index = [] # extracted even placed elements which were values for h5-index
    for i in range(0, len(temp)):
        if not i % 2:
            h5_index.append(temp[i])

    # 10.) iv) h5-median

    h5_median = []
    for i in soup.find_all('span', class_="gs_ibl gsc_mp_anchor"):
        h5_median.append(i.text)
    
    df = pd.DataFrame({'Rank':rank, 'Publication':publication, 'h5-index':h5_index, 'h5-median':h5_median})
    return(df)

pub("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```
