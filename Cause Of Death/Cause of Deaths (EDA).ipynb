### Finding Top diseases from 1990 - 2019

```python
disease = dataset.drop(['Country/Territory', 'Year'], axis = 1)
```

```python
disease
```

```python
Disease = pd.DataFrame(disease.columns)

Disease.columns = ['Disease']
```

```python
deaths = pd.DataFrame(np.sum(disease.values, axis=0), columns=['Deaths'])
```

```python
Disease['Deaths'] = deaths['Deaths']
```

```python
Disease
```

```python
Disease = Disease.sort_values(by=['Deaths'],ascending=False).reset_index().drop('index', axis = 1)

Disease['Deaths (in millions)'] = round(Disease['Deaths']/1000000,2)

Disease = Disease.drop('Deaths', axis = 1)

Disease
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
sns.set(font_scale = 1.7)
ax = sns.barplot(x = 'Deaths (in millions)', y = 'Disease', data = Disease)
plt.xlabel('Deaths (in Millions)', fontsize=30);
plt.ylabel('Diseases', fontsize=30);
plt.title('Top Deaths in Past 30 Years (1990-2019)', fontsize=30)
plt.tick_params(labelsize=20)
ax.bar_label(ax.containers[0])
plt.show()
```

### Top 5 Death causing Diseases are:
1. Cardiovascular Diseases
2. Neoplasms
3. Chronic Respiratory Diseases
4. Lower Respiratory Diseases
5. Neonatal Disorders

##### We can see that Cardiovascular Diseases are the most prominent reason for death and it is almost twice of Neoplasms - the second most death causing disease.

### Top countries with maximum number of deaths over the period of 1990 - 2019

```python
dataset.head(1)
```

```python
country = dataset.drop('Year',axis = 1)
```

```python
country = country.groupby('Country/Territory').sum().reset_index()
```

```python
country['Deaths'] = country.drop('Country/Territory', axis = 1).sum(axis = 1)
```

```python
country
```

```python
top_country = country[['Country/Territory', 'Deaths']].sort_values(by=['Deaths'],ascending=False)[:10]
```

```python
top_country
```

```python
top_country['Deaths (in millions)'] = round(top_country['Deaths']/1000000, 2)

top_country = top_country.drop('Deaths', axis = 1)

top_country.index = range(1, 11)

top_country
```

```python
plt.figure(figsize=(15,8))
sns.set(font_scale = 1.7)
ax = sns.barplot(x = 'Deaths (in millions)', y = 'Country/Territory', data = top_country)
plt.title('Total Deaths in Past 30 Years (1990-2019) vs Top 10 Countries', fontsize=30)
ax.bar_label(ax.containers[0])
plt.show()
```

### From the above bar plot we can see that China has most number of deaths with 265.41 million deaths followed by India with 238.16 million deaths. The United States is at 3rd position but the difference is huge between 2nd and 3rd position, the US has almost 1/3rd deaths occured in India.

The most populated state is China followed by India and United States as of the 2019 population data. Hence, we can see a direction relation with the population of the country.

### Now let's see yearly report for top 3 countries.

```python
sum_disease = dataset
```

```python
sum_disease['Deaths'] = sum_disease.sum(axis = 1)
```

```python
sum_disease = sum_disease.drop(disease.columns, axis = 1)
```

```python
sum_disease
```

```python
top_country_sum_disease = sum_disease[sum_disease['Country/Territory'].isin(["China", "India", "United States"])]
```

```python
top_country_sum_disease.rename(columns = {'Country/Territory':'Country'}, inplace = True)
```

```python
top_country_sum_disease
```

```python
from pandasql import sqldf
```

```python
q = 'select * from top_country_sum_disease where Country == "China"';
china = sqldf(q)

china['Deaths (in millions)'] = round(china['Deaths']/1000000, 2)

china = china.drop('Deaths', axis = 1)

china
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = 'Year', y = 'Deaths (in millions)', data = china)
plt.title('Total Deaths in Past 30 Years (1990-2019) in "CHINA" vs Year', fontsize=25)
plt.show()
```

```python
q = 'select * from top_country_sum_disease where Country == "India"';
india = sqldf(q)

india['Deaths (in millions)'] = round(india['Deaths']/1000000, 2)

india = india.drop('Deaths', axis = 1)

india
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = 'Year', y = 'Deaths (in millions)', data = india)
plt.title('Total Deaths in Past 30 Years (1990-2019) in "INDIA" vs Year', fontsize=25)
plt.show()
```

```python
q = 'select * from top_country_sum_disease where Country == "United States"';
usa = sqldf(q)

usa['Deaths (in millions)'] = round(usa['Deaths']/1000000, 2)

usa = usa.drop('Deaths', axis = 1)

usa
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = 'Year', y = 'Deaths (in millions)', data = usa)
plt.title('Total Deaths in Past 30 Years (1990-2019) in "UNITED STATES" vs Year', fontsize=25)
plt.show()
```

### Now let's see top 3 diseases yearly deaths

```python
yearly_data = dataset.drop('Country/Territory', axis = 1)
```

```python
yearly_data = yearly_data.groupby('Year').sum().reset_index()
```

```python
top_disease_yearly_data = yearly_data[['Year', 'Cardiovascular Diseases', 'Neoplasms', 'Chronic Respiratory Diseases']]
```

```python
df = top_disease_yearly_data[['Cardiovascular Diseases', 'Neoplasms', 'Chronic Respiratory Diseases']].transform(func = lambda x : round(x/1000000, 2))

df.columns = ['Cardiovascular Diseases (in millions)', 'Neoplasms (in millions)', 'Chronic Respiratory Diseases (in millions)']

df['Year'] = top_disease_yearly_data['Year']
```

```python
df
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x = 'Year', y = 'Cardiovascular Diseases (in millions)', data = df)
plt.title('Total Deaths in Past 30 Years (1990-2019) by "Cardiovascular Diseases" vs Year', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x = 'Year', y = 'Neoplasms (in millions)', data = df)
plt.title('Total Deaths in Past 30 Years (1990-2019) by "Neoplasms" vs Year', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x = 'Year', y = 'Chronic Respiratory Diseases (in millions)', data = df)
plt.title('Total Deaths by "Chronic Respiratory Diseases" vs Year', fontsize=25)
plt.show()
```
